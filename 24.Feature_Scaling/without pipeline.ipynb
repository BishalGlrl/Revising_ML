```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df = df[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Survived']]
```

```python
df
```

```python
X = df.drop(['Survived'],axis = 1,inplace = False)
```

```python
Y = df['Survived']
```

```python
X.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,
                                                Y,
                                                test_size=0.2,
                                                random_state=0)
```

```python
transformer1 = ColumnTransformer(transformers=[
    ('trs1',SimpleImputer(),['Age']),
    ('trs2',SimpleImputer(strategy='most_frequent'),['Embarked']),
    ('ohe_sex_embarked',OneHotEncoder(drop = None,sparse_output=False,dtype=np.int32,handle_unknown='ignore'),[['Sex','Embarked']]),
    ('trs3',MinMaxScaler(),slice(0,10))
],remainder='passthrough')
```

```python
X_train_imputed = transformer1.fit_transform(X_train)
```
