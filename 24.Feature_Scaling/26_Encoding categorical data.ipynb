## There are two types of categorical data. 1) Nominal: This is random data with no meaningful order.For eg: (Male,female),(red,green,blue),(yes,no)
## Next one is 2) Ordinal(Ordered) data :This is arranged data.For eg (distinction,first division,second division),(High,medium,low),(Neutral,satisfied,not satisfied),(Bachlor,masters,phd)

<h2 style = "color:red;">Encoding is the process where values of categorical data is converted into numeric form suitable for analysis by machine learning algorithm.<br>Here we have two types of encoding <b> 1)Ordinal Encoding</b>(for ordinal data) and <b>2) one hot encoding</b>(For Nominal data)and <b> 3) Binary encoding</b>b for yes no data i.e. Spam,not spam.Drink dont drink or 0,1 data</h2>

# 1) ORDINAL ENCODING

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.sample(5)
```

### In above, Age ,Gender are Nomial.Review and Education are Ordinal.Purchased is Label

### Now we have to perform one hot encoding in Nominal Attribute and Ordinal Encoding in OrdinaL attribute and Label encoding in Purchase

```python
from sklearn.model_selection import train_test_split
```

### FOR ORDINAL ATTRIBUTE ,performing ordinal Encoding

```python
X = df.iloc[:, [2,3]]
Y = df.iloc[:,-1]
```

```python
X,Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,
                                                Y,
                                                test_size=0.2,
                                                random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
categories = [['Poor','Average','Good'],['School','UG','PG']]
oe = OrdinalEncoder(categories=categories)
```

<h2 style = 'color:green;'>Here in "categories = [['Poor','Average','Good'],['School','UG','PG']]" we are saying that assign [0,1,2] to [['Poor','Average','Good'] and to ['School','UG','PG'] as Poor = 0,Average = 1, and Good =2.If we dont pass the caregories in "oe = OrdinalEncoder(categories=categories) then the number would be assigned randomly i.e. 0 to Average and 1 to Good and 2 to Poor but we dont want to randomly assign it"</h2>

```python
X_train_encoded = oe.fit_transform(X_train)
```

```python
X_train_encoded
```

## Now for LabelEncoder.We always apply LabelEncoder to the class label of the dataframe and not to the input feature.Here we cannot decide the value to be assigned to the categories as in above OrdinalEncoder.LabelEncoder itself decides the value to the label.So it may be 0 for "YES" and 1 for "NO"

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
Y_train_encoded = le.fit_transform(Y_train)
Y_test_encoded = le.transform(Y_test)
```
