## PIpeline chain together multiple steps so that the output of each step is used as input to next step.Pipelining makes it easy to apply the same preprocessing to train and test.Following are the tasks in pipelining.
1) Handling missing value(using SimpleImputer) <br>2) Using OneHotEncoder <br>3) Scaling <br>4) Feature selection <br>5) Train the model

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head(1)
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head(1)
```

```python
X = df.drop(columns=['Survived'],inplace = False)
```

```python
Y = df['Survived']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,
                                                Y,
                                                test_size=0.2,
                                                random_state=0)
```

```python
X_train
```

```python
# column_index = 6
column_name = X_train.columns[6]
X_train = pd.get_dummies(X_train,columns=[column_name],dtype=int)
```

```python
column_name = X_train.columns[1]
X_train = pd.get_dummies(X_train,columns = [column_name],dtype= int)
```

```python
X_train
```

```python
X_train.isnull().sum()
```

```python
# To handle null value
transformer1 = ColumnTransformer(transformers=[
    ('trs1',SimpleImputer(),[2]),
    ('trs2',SimpleImputer(strategy = 'most_frequent'),[6])
],remainder='passthrough')
```

```python
#Performing One Hot Encoding
# transformer2 = ColumnTransformer(transformers=[
#     ('trs1',OneHotEncoder(drop = None,sparse_output = False,dtype = np.int32,handle_unknown = 'ignore'),[1,6])
# ],remainder='passthrough')
```

```python
# Performing scaling
transformer3 = ColumnTransformer(transformers = [
    ('trs1',MinMaxScaler(),slice(0,10))
],remainder='passthrough')
```

```python
#Feature selection
transformer4 = SelectKBest(score_func=chi2,k=8)
```

```python
# Training the model
transformer5 = DecisionTreeClassifier()
```

# NOT LAST STEP IS TO DO PIPELINING

```python
pipe = Pipeline([
    ('trs1',transformer1),
    # ('trs2',transformer2),
    ('trs3',transformer3),
    ('trs4',transformer4),
    ('trs5',transformer5)
])
```

```python
p = pipe.fit(X_train,Y_train)
```
