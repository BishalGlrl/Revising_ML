## Binning, in very simple terms, is the process of grouping continuous data into discrete categories or "bins." This is done to simplify data analysis and make patterns or trends more apparent.

## Example:
## Imagine you have a list of ages:

## [5, 12, 18, 23, 30, 45, 60]

## Binning these ages might involve grouping them into categories like:

## 0-18 years (young)
## 19-35 years (adult)
## 36-60 years (middle-aged)

<h2 style = 'color:red;'>In this one,we will be learning about Unsupervised(Known width(Uniform binning),Known frequency(Quantile binning),Kmean binning) and custom binning.We will be reading about Supervised binning later as it is not that widely used</h2>

<B>Here in Known width,for eg we have a data [5, 12, 18, 23, 30, 45, 60],then we will be arranging the data in a equal size bin with equal interval. like Bin 1 will contain data from [0-10],Bin 2 from [10-20],Bin 3 from [30-40 ] and so on</B>

<B>In known frequency, each bin will be splitted in term of quartile.For eg if quartile is 10 then Bin1 will be containing the first 0-10% of data that may be from 1-7 and Bin2 will be containing from (10-20)% that may be from 7-10,Bin3 will be containing from (20-30)% that may be from 11-21 and so on.<br>Simple ma vannu parda percentage of data ko hisab ma data lai bins ma split garxa  </B>

# FOR BINNING

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
X = df.drop('Survived',axis = 1)
Y = df['Survived']
```

```python
X
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,
                                                Y,
                                                test_size=0.2,
                                                random_state=0)
```

```python
X_train
```

```python
# St = SimpleImputer(strategy='most_frequent')
ctrans = ColumnTransformer(transformers=[
    ('trf1',SimpleImputer(),[0]),
    ('trf2',SimpleImputer(strategy='most_frequent'),[1])
],remainder='passthrough')
```

```python
X_train = ctrans.fit_transform(X_train)
X_test = ctrans.transform(X_test)
```

```python
X_train=pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
```

```python
X_train
```

```python
X_test
```

```python
X_train = X_train.rename(columns={0:'Age',1:'Fare'})
X_test = X_test.rename(columns = {0:'Age',1:'Fare'})
```

```python
trf = ColumnTransformer(transformers=[
    ('trf1',KBinsDiscretizer(n_bins=10, encode="ordinal",strategy='quantile'),[0]),
    ('trf2',KBinsDiscretizer(n_bins = 15,encode = "ordinal",strategy = 'uniform'),[1])
], remainder='passthrough')
```

```python
X_train_trf = trf.fit_transform(X_train)  # fits the transformers to X_train and transforms it, producing X_train_trf.
X_test_trf= trf.transform(X_test) #transform applies the already fitted transformer to  X_test
```

```python
X_train_trf = pd.DataFrame(X_train_trf)
```

```python
X_train_trf = X_train_trf.rename(columns = {0:'Age',1:'Fare'})
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train_trf,Y_train)
```
