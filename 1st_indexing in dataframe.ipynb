<h2>
Lets understand difference between loc and iloc.
Use loc when you want to select data based on labels (row and column names).df.loc[3]  # Select row with index label 3,df.loc[df['A'] > 5]  # Select rows where column 'A' is greater than 5 <br><br>
Use iloc when you want to select data based on the integer positions (0-based index).df.iloc[1:4, 0:2]  # Select rows from position 1 to 3 and columns from position 0 to 1

</h2>

<h2 style = 'color:purple;'>Note that :2 means 0 and 1, 2: mean 2,3,4... and 2:4 means 2,3  </h2>

<h2 style = 'color : blue;'>Now in indexing,when we do iloc[1,2] then it means iloc[row,column].If we do iloc[:,'col'] then i means selecting all the row(:) of column (col)<h2>

<h2 style='color:red;'>1. Selecting Specific Rows and Columns</h2>

### Select the first two rows and the first three columns:

```python
df.iloc[0:2 , 0:3]
```

<h2 style = 'color :red;'>2)Selecting All Rows for Specific Columns</h2>

### Select all rows and the first two columns

```python
df.iloc[:,0:2]
```

<h2 style = 'color :red;'>3)Selecting Rows Based on Conditions</h2>

### Select rows where age is greater than 25:

```python
df[df['Age']>30]
```

<h2 style ='color:red ;'> 4) Selecting column based on condition  </h2>

```python
df.iloc[: , :2] ## keep all the rows but return the column till 0,1
```

<h2 style = 'color :red;'>4)Selecting Columns in Non-Continuous Order</h2>

### Select the first and third columns and all row:

```python
df.iloc[: , [0,2]]
```

<h2 style = 'color :red;'>5)Using Negative Indices</h2>

```python
df.iloc[:-3 , :-2] # Reduce the last three rows and last two columns
```

### Select the last two rows and the last three columns

```python
df.iloc[-3:,-2 :]
```

<h2 style = 'color :red;'>6)Using Step in Slicing</h2>

### Select every other column

```python
df.iloc[:,::2] 
```

### Select every other row

```python
df.iloc[::2, :]
```
