# 4. HeatMap (Categorical - Categorical)

```python
titanic.head()
```

### Now from above,let me find out survived people in PClass

```python
pd.crosstab(titanic['Survived'],titanic['Pclass'])
```

### Now from above we can make out own heat map

```python
sns.heatmap(pd.crosstab(titanic['Survived'],titanic['Pclass']))
```

### Here light color means death of people is low.It is showing that survived people in class 1 is moderate but class 2 is high
### Also not survived people in class 3 is low and moderate in class 2 but high in class 1

### There can be more people in some class from where more people may have died.And due to less people in some class,less
### people might have died.So for accurate we have do do this comparision by percentage as

```python
titanic.groupby('Pclass').mean()['Survived']
# This gives the average survival rate for each passenger class.
```

### This error is due to non numeric value in the data

```python
titanic.select_dtypes('number')
```

```python
titanic.groupby('Pclass').mean(numeric_only=True)['Survived']
# This gives the average survival rate for each passenger class.
```

### The above code groups the classes of Pclass i.e. 1,2 and 3 .And there for each classes,it calculated the survival rate i.e. where it is 1
### and return it as a means.for eg the class of 1 has survival rate of 62.96%
### This at first groups the class of only 1.And then look the value of survived there and the calculated the mean as
### 1+0+0+0+1+1+0+1........+n/n

```python
titanic.groupby('Pclass')['Survived'].mean(numeric_only=True)
# This gives the average survival rate for each passenger class.
```

```python
titanic.groupby('Pclass')['Survived'].mean(numeric_only=True).plot(kind = 'bar')
```
