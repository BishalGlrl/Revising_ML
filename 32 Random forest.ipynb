## Just like a forest contains a multiple trees to be dense forest,Random forest also contains a multiple decision treees where each tree predict their own decision and random forest makes prediction based on those decision .<br><br>Inside random forest there multiple decision trees and each decision trees are not trained with whole dataset but only with random subset of data tuples and features.<br><br>For ecample if a dataset contains 1000 rows and 10 features and there are three decision trees in random forest then 1 decision tree can be trained with randomly generated 700 rows and 4 columns,another with the completely random 700 rows and 4 column which is completely different from the data fed to first decision tree and third decision tree also be trained with completely random 700 data and 3 coolumns or features

### Random forest makes prediction in 2 forms: <br><br> 1) Classification: For classification probelms,each decision tree might provide a different predictions and random forest find the most occuring decision among all decision tree and give prediction.<br>2) Regression : Instead of making prediction based on the most ocured decision from decision tree as in classification,here in regression random forest makes prediction by averaging the decision from each decision tree.<br><br>For eg:When you want to predict if it will rain tomorrow, you ask all the trees in the forest for their predictions. Each tree gives you its answer: "Yes, it will rain" or "No, it won't rain."<br><br>For Classification (e.g., Will it rain? ans will be either 1 or 0 i.e. yes or no): The Random Forest collects the predictions from all the trees and chooses the one that most trees agree on. If most trees say "Yes," or 1 the final prediction is "Yes." or 1<br><br>For Regression (e.g., How much rain?): If the task was to predict the amount of rain(Regression ma categorical prediction hunna), the Random Forest would average the predictions from all the trees.

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
import pandas as pd
import numpy as np
```

```python
data
```

```python
X = data['data']
Y = data['target']
```

```python
X = pd.DataFrame(X)
Y = pd.DataFrame(Y)
```

```python
X.isnull().sum()
```

```python
Y.isnull().sum()
```

```python
Y.rename(columns={0 :'target'},inplace = True)
```

```python
flower = []
for val in Y.target:
    if val == 0:
        flower.append('setosa')
    if val ==1:
        flower.append('versicolor')
    if val ==2:
        flower.append('virginica')
print(pd.DataFrame(flower,columns=Y.columns))
```

```python
temp = ['sepal length','sepal width','petal length', 'petal width' ]
X.columns = temp
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=43)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100) ## 100 ota decision tree banauna vaneko ho yesle.We can adjust it to anything
```

```python
rfc.fit(X_train,Y_train)
```

```python
Y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
confusion_matrix(Y_test,Y_pred)
```
