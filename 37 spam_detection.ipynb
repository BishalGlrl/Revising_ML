```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('spam.csv')
```

```python
### There are two unique class in df['category']
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
temp = Le.fit_transform(df['Category'])
df['category'] = pd.DataFrame(temp)
# temp = pd.get_dummies(df['Category'],dtype=np.int_)
```

```python
df.drop('Category',axis=1,inplace = True)
```

```python
df
```

## Now lets convert Message to numeric using count vectorizer method

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X = df.drop('category',axis=1)
Y = df['category']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=43)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.pipeline import Pipeline
# Label = Pipeline(steps=[
#     ('pipe1',LabelEncoder())
# ])
## Label encoder is not designed to work with pipeline
Count = Pipeline(steps=[
    ('pipe2',CountVectorizer())
])

MN = Pipeline(steps=[
    ('pipe3',MultinomialNB())
])

from sklearn.compose import ColumnTransformer
features = [0]
preprocessed_data = ColumnTransformer(transformers=[
    # ('trf1', Label,0),
    ('trf2',Count,'Message')
],remainder='passthrough')
```

```python
model = MultinomialNB()
```

```python
clf = Pipeline(steps=[
    ('tr1',preprocessed_data),
    ('tr2',model)
])
```

```python
clf.fit(X_train,Y_train)
```

#### Above are the vectors made for each columns which are as

```python
Y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```

```python
from sklearn.metrics import confusion_matrix
conf = confusion_matrix(Y_test,Y_pred)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(conf,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
