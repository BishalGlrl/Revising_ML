### At first lets understand what is coefficient in dataset.Coefficient(B) is the total weight or effect a column have on target label.Each column have different coefficient i.e. different effect on target label

### Lasso and Ridge regression are techniques used in linear regression to prevent overfitting by adding a penalty to the model's complexity i.e. adding another parameter to the model MSE

## While applying Lasso regression,it can make a coefficient of some features to zero i.e. removes the features from dataset .so it is not used when features are hardly dependent on target <br> But Ridge regression doesnot shrink coeefficient of column to zero.Hence we can use it in most cases

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,0:3]
Y = df['Outcome'].astype(float)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=43)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
```

### Using LinearRegression

```python
from sklearn.metrics import r2_score
LR = LinearRegression()
LR.fit(X_train,Y_train)
Y_pred = LR.predict(X_test)
r2_score(Y_test,Y_pred)
```

### Using lasso

```python
La = Lasso(alpha=0.1)
La.fit(X_train,Y_train)
Y_pred = La.predict(X_test)
r2_score(Y_test,Y_pred)
```

### Using Ridge

```python
Ri = Ridge(alpha=0.1)
Ri.fit(X_train,Y_train)
Y_pred = Ri.predict(X_test)
r2_score(Y_test,Y_pred)
```

## Now lets use hypertuning to find out which value of alpha can be a best suit

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha' : [0.1,0.5,1,1.2,4,56,100]}
gsc = GridSearchCV(estimator=Lasso(),
                  param_grid = param_grid,
                  n_jobs=-1,
                  cv=5,
                  verbose=2
                   
                   )
```

```python
gsc.fit(X_train,Y_train)
```

```python
gsc.best_params_
```

#### So alpha value of 0.1 is best parameters for it

```python
best_model_lasso = gsc.best_params_
```

```python
y_p = gsc.predict(X_test)
r2_score(Y_test,y_p)
```

## For Ridge Regression

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha' : [0.1,0.5,1,1.2,4,56,100]}
gc = GridSearchCV(estimator=Ridge(),
                  param_grid = param_grid,
                  n_jobs=-1,
                  cv=5,
                  verbose=2
                   
                   )
```

```python
gc.fit(X_train,Y_train)
```

```python
gc.best_params_
```

#### In Ridge,alpha value of 100 is best parameters

```python
y = gc.predict(X_test)
r2_score(Y_test,y)
```
