```python
df = pd.read_csv('placement.csv')
```

```python
df.head(3)
```

```python
import matplotlib.pyplot as plt
figure,ax = plt.subplots(figsize=(10,5))
ax.set_title("Distribution of data")
sns.scatterplot(x=df['cgpa'],y=df['package'],ax=ax)
plt.show()
```

## Here it is linear relation so we can apply linear regression

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,0:1]
Y = df.iloc[:,-1:]
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=43)
```

```python
X_train.shape , X_test.shape
```

```python
X_test
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
lr.predict(X_test.iloc[0])
```

### Model was trained with dataframe so it is expecting 2D input i.e. Dataframe containing both row and column but we provide a series i.e. 1D input,so we have to transform it to 2D by function reshape()

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

## For equation y = mx+c, we will be finding the value of m and c now

```python
slope = lr.coef_
slope
```

```python
c= lr.intercept_
c
```

### Now y will be prediction made by the model based on the given input column x i.e y = mx+c
