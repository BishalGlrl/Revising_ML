## This algorithm is used in unsupervised learning where we have only an input labels but we dont know the target value

<h2 style = 'color:purple;'>Here while doing K mean clusturing,keep in mind centroid,number of clusters and SSE(Sum Square error)</h2>

## Steps :<br>1) Choose a number of clusters<br>2) Randomly selects centroid points i.e. centroid of clusters thus formed.If you have selected n clusters,then there will be 'n' centroid points.<br>3) Calculates the distance of each data points from each centroid and assign it to a nearest centroid<br>4) After all the data points are assigned to a centroid,then update the value of centroid with respect to all the datapoints.<br>5) Find SSE

## SSE(Also called as inertia) :<br> For each  clusters(K),we have certain inertia i.e. k1=SSE1,K2=SSE2,K3=SSE3.....KN=SSEN.This inertia value gets minimize as we increase the number of clusters.So we have to choose the number clusters such that value of SSE will be zero at that cluster.<br>The best way is to draw a line in graph with k respect to SSE.
