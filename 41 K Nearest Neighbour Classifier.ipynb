### KNN classifier plots the training data points in the graph.And when a new test data point is passed to the KNN classifier then it will be finding the "n" nearest neighbour from a test data point in the graph where training data points were plotted and give the prediction based on those neighbours.<br><br>KNN gives prediction based on Classification or Regression problem.<br><br> In the classification problem, the class labels of K-nearest neighbors to test data points are determined by performing majority voting. The class with the most occurrences among the neighbors becomes the predicted class for the test data point.<br><br>In the regression problem, the class label is calculated by taking average of the values(as there will be no class label in regression) of K nearest neighbors to test data points. The calculated average value becomes the predicted output for the test data point.

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
data = load_iris()
X = data['data']
Y = data['target']
```

```python
data
```

```python
data.feature_names
```

```python
X = pd.DataFrame(X,columns=data.feature_names)
```

```python
Y =pd.DataFrame(Y,columns = ['target'])
```

```python
Y 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=43)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier(n_neighbors=9)
```

```python
kn.fit(X_train,Y_train)
```

```python
Y_pred = kn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(Y_test,Y_pred)
```

### Our model gives an score of 96%

```python
confusion_matrix(Y_test,Y_pred)
```

```python
conf = confusion_matrix(Y_test,Y_pred)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(conf,annot= True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

### Lets perform Hyperparameter tuning now

```python
param_grid = {
    'leaf_size' : [10,20,30,40,50,60,70],
    'n_neighbors' : [5,10,40,60,55]
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
gsc = GridSearchCV( estimator = kn,
                    param_grid = param_grid,
                    n_jobs=-1,
                    cv=5,
                    verbose=2)
```

```python
gsc.fit(X_train,Y_train)
```

```python
gsc.best_params_
```

```python
gsc.best_score_
```

```python
y = gsc.predict(X_test)
accuracy_score(Y_test,y)
```
