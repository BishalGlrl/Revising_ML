## A Support Vector Machine (SVM) is a supervised machine learning algorithm primarily used for classification tasks and in some cases regression too.<br><br>The main idea of SVM is to find the best line or boundary (called a hyperplane) that separates data into different classes.In dog class classification it draws a hyperplane such that it can easily distinguish between dog and cat.For 1D features,it draws line,for 2D it draws plane and for n dimesniona it draws hyperplane<br><br>SVM focuses on the most important and close data points, called support vectors, to create this boundary(hyperplane). These support vectors help define the hyperplane that best separates the classes and are close to the hyperplane.Hyperplane is drawn on the basis of support vector and small change in position of SV can cause hyperplane to be highly displaced<br><br>The hyperplane is drawn such that it has the wide margin from the closest point(support vector).Wide margin assures that even if data points moves a little bit from its original position,the line(hyperparameters) still correctly separates the groups.If hyperparameter have a narrow margin then If the dots move slightly, the line might no longer separate the groups correctly.

<img src = 'support-vector-machine-algorithm.png'/>

## In above,our dataset have two groups and a hyperplane(line) is drawn such that it separates the points of two group from the middle.The line drawn depends on the support vectors(nearest point to the line).Line is drawn acctoring to these SV.Small change in position of SV can cause line to displace its position

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
df = load_iris()
```

```python
df
```

```python
X = df['data']
Y = df['target']
```

```python
X = pd.DataFrame(X,columns=['sepal_length','sepal_width','petal_length','petal_width'])
Y = pd.DataFrame(Y,columns=['targets'])
```

```python
df.target_names ##It is provided in above load_iris data
```

```python
flower = []
'''
for index,row in Y.iterrows():
    print(f'{index} {row}') ## Index will be giving index of a row and row will be giving value inside a row
    for value in row :
        if value == 0:
            flower.append('setosa')
        elif value == 1:
            flower.append('versicolor')
        elif value == 2 :
            flower.append('virginica')
flower = pd.DataFrame(flower)
'''

for val in Y.targets:
    if val == 0:
            flower.append('setosa')
    elif val == 1:
        flower.append('versicolor')
    elif val == 2 :
        flower.append('virginica')
flower = pd.DataFrame(flower)
```

### Here when we apply SVM ,then it can siimply draw a hyperplane in between orange dot and blue '+' symbols and can provide a clear distinction between them

```python
from sklearn.svm import SVC
svc = SVC()
```

```python
from sklearn.model_selection import train_test_split
columns = dataframe.iloc[:,-2:]
X = dataframe.drop(columns,axis = 1)
Y = dataframe.iloc[:,-2:-1] ## THis will extract only the second last column i.e. -2:-1 or second_last: last means means only selecting second last 
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=43)
dataframe.shape , X_train.shape,X_test.shape
```

```python
svc.fit(X_train,Y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(Y_test,y_pred)
```

```python
confusion_matrix(Y_test,y_pred)
```

### Lets convert above matrix in a form which we can visualize goodly

```python
import seaborn as sns
conf = confusion_matrix(Y_test,y_pred)
sns.heatmap(conf,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion matrix')
plt.show()
```

## From above we can analyzed that '0' class that were 0 were predicted 2,'2' class were predicted 1 and '7' were predicted 2 also,for '1' class,8 class were predicted 1 which were actual 1 and no wrong prediction where made and same for 0
