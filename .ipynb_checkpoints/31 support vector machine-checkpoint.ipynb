## A Support Vector Machine (SVM) is a supervised machine learning algorithm primarily used for classification tasks and in some cases regression too.<br><br>The main idea of SVM is to find the best line or boundary (called a hyperplane) that separates data into different classes.For 1D features,it draws line,for 2D it draws plane and for n dimesniona it draws hyperplane<br><br>SVM focuses on the most important and close data points, called support vectors, to create this boundary. These support vectors help define the hyperplane that best separates the classes and are close to the hyperplane.Hyperplane is drawn on the basis of support vector and small change in position of SV can cause hyperplane to be highly displaced<br><br>The hyperplane is drawn such that it has the wide margin from the closest point(support vector).Wide margin assures that even if data points moves a little bit from its original position,the line(hyperparameters) still correctly separates the groups.If hyperparameter have a narrow margin then If the dots move slightly, the line might no longer separate the groups correctly.

## In above,our dataset have two groups and a hyperplane(line) is drawn such that it separates the points of two group from the middle.The line drawn depends on the support vectors(nearest point to the line).Line is drawn acctoring to these SV.Small change in position of SV can cause line to displace its position

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
df = load_iris()
```

```python
df
```

```python
X = df['data']
Y = df['target']
```

```python
X = pd.DataFrame(X,columns=['sepal_length','sepal_width','petal_length','petal_width'])
Y = pd.DataFrame(Y,columns=['targets'])
```

```python
df.target_names ##It is provided in above load_iris data
```

```python
X
```

```python
flower = []
'''
for index,row in Y.iterrows():
    print(f'{index} {row}') ## Index will be giving index of a row and row will be giving value inside a row
    for value in row :
        if value == 0:
            flower.append('setosa')
        elif value == 1:
            flower.append('versicolor')
        elif value == 2 :
            flower.append('virginica')
flower = pd.DataFrame(flower)
'''

for val in Y.targets:
    if val == 0:
            flower.append('setosa')
    elif val == 1:
        flower.append('versicolor')
    elif val == 2 :
        flower.append('virginica')
flower = pd.DataFrame(flower)
```

```python
flower.rename(columns={0:'flower'},inplace = True)
```

```python
Y['flower'] = flower['flower']
```
