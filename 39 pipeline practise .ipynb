```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
df['size'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
LE = LabelEncoder()
df['Sex'] = pd.DataFrame(LE.fit_transform(df['sex']))
```

```python
df.drop('sex',axis=1,inplace=True)
```

```python
df['Smoker'] = pd.DataFrame(LE.fit_transform(df['smoker']))
df['Time'] = pd.DataFrame(LE.fit_transform(df['time']))
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('tip',axis=1)
Y = df['tip']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=43)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
rfc = RandomForestClassifier(n_estimators=20)
preprocessing = ColumnTransformer(transformers=[
    ('trs1',OrdinalEncoder(),[1,2,4]),
    ('trs2',OneHotEncoder(sparse_output=False,dtype=np.int_),[3])
],remainder='passthrough')

from sklearn.pipeline import Pipeline
pp = Pipeline(steps=[
    ('pipe1',preprocessing),
    ('pipe2',LinearRegression())
])


```

```python
X_train
```

```python
pp.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score,r2_score
Y_Pred = pp.predict(X_test)
r2_score(Y_test,Y_Pred)
```
